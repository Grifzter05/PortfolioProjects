## Submuestreo

Construimos una muestra a partir de nuestro marco de datos que contenga una distribuciÃ³n similar de las transacciones 
fraudulentas y normales

Tomando en cuenta que el numero de transacciones fraudulentas consta de 492 observaciones.

```python
legit_sample = legit.sample(n = 492)
```

```python
new_dataset = pd.concat([legit_sample,fraud], axis=0)
new_dataset
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Dividimos los datos en caracterÃ­sticas y objetivos

```python
x = new_dataset.drop(columns = 'Class', axis = 1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Dividimos los datos en datos de prueba y datos de entrenamiento (test & train)

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.2, stratify = y, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# Modelo de entrenamiento

## RegresiÃ³n logaritmica

```python
model = LogisticRegression(max_iter=1000) #Aumentamos el numero de iteraciones para que se alcance una mayor convergencia
```

```python
#Entrenando el modelo logaritmico con datos de entrenamiento 
model.fit(x_train, y_train)
```

# EvaluaciÃ³n del modelo

## Puntaje de precisiÃ³n del modelo

```python
# PrecisiÃ³n de los datos de entrenamiento 
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print ('Accuracy on training data: ', training_data_accuracy)
```

```python
#Obtenemos un buen puntaje, lo que quiere decir que las predicciones seran buenas
```

```python
# PrecisiÃ³n de los datos de prueba
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print ('Accuracy score on test data: ', test_data_accuracy)
```
